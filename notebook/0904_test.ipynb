```python
import pandas as pd

xml = pd.read_table("../data/grant_bib_2012/ipgb20120103.xml", header=None)
xml.head(50)
```

```python
mask = xml[0].str.contains('<publication-reference>')

add = [False, False, False]
add_df = pd.Series(add)

mask_pub = pd.concat([add_df, mask]).reset_index(drop = True)
mask_pub
```

```python
xml_pub = xml[0].where(mask_pub).dropna(how = all)
xml_pub
```

```python
mask_app = pd.concat([add_df, xml[0].str.contains('<application-reference appl-type="utility">')]).reset_index(drop = True)
xml_app = xml[0].where(mask_app).dropna(how = all)
xml_app
```

```python
xml_pub_num = xml_pub.str[12:20]
xml_pub_df = pd.DataFrame(xml_pub_num)
xml_pub_df[1] = xml_pub_num.index.values.tolist()
xml_pub_df = xml_pub_df.reset_index(drop = True)
xml_pub_df
```

```python
xml_app_num = xml_app.str[12:20]
xml_app_df = pd.DataFrame(xml_app_num)
xml_app_df[1] = xml_app_num.index.values.tolist()
xml_app_df[1] = xml_app_df[1] -8
xml_app_df
```

```python
app_pub_df = pd.merge(xml_app_df, xml_pub_df, on = 1)
app_pub_df = app_pub_df.rename(columns={'0_x': 'app_num', '0_y': 'pub_num'})
app_pub_df
```

```python
app_pub_df['pub_num'] = app_pub_df['pub_num'].str[1:8]
```
