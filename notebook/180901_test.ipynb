```python
import re
def get_app_id(xml):
    flg = 0
    for row in xml:
        if row.startswith('<doc-number>'):
            if flg == 1:
                return re.match('<doc-number>(.*)</doc-number>', row).group(1)
            flg  += 1
```

```python
app_df['application_id3'] = app_df.patent_data.map(lambda x:get_app_id(x))
```

```python
app_df.head()
```

```python
app_df.application_id == app_df.application_id3
```

```python
app_df_true = app_df[app_df.application_id == app_df.application_id3]
```
