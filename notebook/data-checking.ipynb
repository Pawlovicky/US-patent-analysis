A new <?xml version="1.0" encoding="UTF-8"?> line causes the error.

```python
import re
import lxml.etree

re_strip_pi = re.compile('<\?xml [^?>]+\?>', re.M)
data = '<root>' + open('../data/ipg150120.xml', 'rb').read().decode("ascii") + '</root>'


match = re_strip_pi.search(data)
data = re_strip_pi.sub('', data)

test = match.group() + data

re_strip_doc = re.compile('<!DOCTYPE [^?>]+>', re.M)
match = re_strip_doc.search(data)
data = re_strip_doc.sub('', data)

test = match.group() + data

# tree = lxml.etree.fromstring(test)
doc = xmltodict.parse(test)
```
