```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
path_app = "../data/app_bib_2017"
path_grant = "../data/grant_bib_2012"

Files_app = glob.glob(os.path.join(path_app,"*.xml"))
Files_grant = glob.glob(os.path.join(path_grant,"*.xml"))
```

```python
Files_app
```

```python
Files_grant
```

```python
list_app = []

col_names = [ 'c{0:02d}'.format(i) for i in range(3) ]

for file_app in Files_app:
    f_app = pd.read_table(file_app, header=None, names=col_names )
    list_app.append(f_app.as_matrix())
```

```python
np_array_app = np.vstack(list_app)
big_frame = pd.DataFrame(np_array_app)
```

```python
big_frame
```

```python
big_frame_app = big_frame.drop([1,2], axis = 1)
```

```python
big_frame_app
```

```python
list_grant = []

col_names = [ 'c{0:02d}'.format(i) for i in range(2) ]

for file_grant in Files_grant:
    f_grant = pd.read_table(file_grant, header=None, names=col_names )
    list_grant.append(f_grant.as_matrix())
```

```python
np_array_grant = np.vstack(list_grant)
Big_frame = pd.DataFrame(np_array_grant)
```

```python
Big_frame[~Big_frame[1].isnull()]
```

```python
big_frame_grant = Big_frame.drop(1, axis = 1)
```

```python
big_frame_grant
```

```python
mask = big_frame_grant[0].str.contains('<publication-reference>')

add = [False, False, False]
add_df = pd.Series(add)

mask_pub = pd.concat([add_df, mask]).reset_index(drop = True)
mask_pub
```

```python
xml_pub = big_frame_grant[0].where(mask_pub).dropna(how = all)
xml_pub
```

```python
xml_pub_num = xml_pub.str[12:20]
xml_pub_df = pd.DataFrame(xml_pub_num)
xml_pub_df[1] = xml_pub_num.index.values.tolist()
xml_pub_df = xml_pub_df.reset_index(drop = True)
xml_pub_df
```

```python
xml_pub_df_pat = xml_pub_df[~xml_pub_df[0].str.startswith('D') & ~xml_pub_df[0].str.startswith('P') & ~xml_pub_df[0].str.startswith('R')]
```

```python
xml_pub_df_pat
```

```python
xml_pub_df_pat[0] = xml_pub_df_pat[0].str[1:8]
```

```python
xml_pub_df_pat
```

```python
xml_pub_df_pat = xml_pub_df_pat.rename(columns={0: 'parsed'}).drop(1, axis = 1)
```

```python
xml_pub_df_pat
```
